Ngày đầu tiên (ngày 0) có n cá thể ở mức 1, Ở mỗi ngày tiếp theo, mỗi cá thể mức i sinh ra i cá thể mức 1, các cá thể mới sinh sẽ sinh sôi, phát triển từ ngày hôm sau. Bản thân các cá thể mức i phát triển thành mức i+1 và chu kỳ phát triển trong ngày chấm dứt. Hãy xác định sau k ngày trong nước biển có bao nhiêu cá thể.

Dữ liệu: Vào từ thiết bị nhập chuẩn gồm một dòng chứa 2 số nguyên n và k (1 ≤ n ≤ 1000, 1 ≤ k ≤ 10^17 ).

#Algorithm design 


+Tạo 1 hàm fibo(x),Với x>2

+Gán x0=1,x1=1,xn=1.Ta duyệt từ 2 đến x-1 và tính chuỗi phép tính,xn=x0+x1,x0=x1,x1=xn.Sau khi hoàn thành vòng lặp trả về xn.

+Kết quả cuối cùng sẽ là (n*fibo(x)%(10^9+7))

#Độ phức tạp của bài toán 

Do ta duyệ từ 2 đến k-1 để thức hiện 2 phép toán cơ bản.Vì vậy độ phức tạp của thuật toán trên là 0(N)

#Programming

```python
n,k=map(int,input().split())
def fibonacci(x):
  if x<2:
     return 1
  else: 
      x0=1
      x1=1
      xn=1
      for i in range(2,x):
           xn=x0+x1
           x0=x1
           x1=xn
      return xn
print((n * fibonacci(2*k + 1)) % (10**9 + 7))

```
